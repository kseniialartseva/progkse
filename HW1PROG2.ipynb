```python
import re
import gensim
import logging
import nltk.data
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from gensim.models import word2vec

import warnings
warnings.filterwarnings('ignore')

from pymystem3 import Mystem
import pymorphy2
import codecs
```

```python
fileObj = codecs.open("catcherintherye.txt", "r", "utf_8_sig" )
text = fileObj.read()
fileObj.close()
```

```python
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(text)
print(sentences)
```

```python
print(text)
```

```python
marks = '''!()-[]{};?@#$%:'"\,./^&amp;*_'''

for sent in sentences:
    for x in text:  
        if x in marks:  
            sent = sent.replace(x, "")

for sent in sentences:
    sent = sent.lower()
    
print(sentences)
    
```

```python
m = Mystem()
mass_of_lemmas = []
for sent in sentences:
    mass_of_lemmas.append(m.lemmatize(sent))
```

```python
print(mass_of_lemmas)
```

```python
fw = codecs.open("catcherintherye.txt", "w", "utf_8_sig" )
for lem in mass_of_lemmas:
    fw.write(''.join(lem))
fw.close()
```
